# (二) PDF表格處理

```python
import pdfplumber
import pandas as pd

with pdfplumber.open("11_R-30iA_HandlingTool基本操作教育訓練教材(NEW).pdf") as pdf:
    all_tables = []  # 初始化一個空的列表以儲存所有頁面的表格資料

    # 迭代PDF檔案中的每一頁
    for page in pdf.pages:
        table = page.extract_table()  # 從當前頁面提取表格資料
        if table:  # 如果當前頁面有表格資料
            df = pd.DataFrame(table[1:], columns=table[0])  # 創建資料框架，表頭為表格的第一行
            df = df.replace(r'\n', ' ', regex=True)  # 替換資料框架中的換行符為空格
            all_tables.append(df)  # 將處理過的資料框架加入到all_tables列表中
```

```python
import pdfplumber
import pandas as pd

def extract_tables_from_pdf(pdf_path):

    all_tables = []  # 初始化一個空的列表以儲存所有頁面的表格資料
    with pdfplumber.open(pdf_path) as pdf:
        # 迭代PDF檔案中的每一頁
        for page_num, page in enumerate(pdf.pages):
            table = page.extract_table()  # 從當前頁面提取表格資料
            if table:  # 如果當前頁面有表格資料
                df = pd.DataFrame(table[1:], columns=table[0])  # 創建資料框架，表頭為表格的第一行
                df = df.replace(r'\n', ' ', regex=True)  # 替換資料框架中的換行符為空格
                all_tables.append({f"Page_{page_num+1}": df.to_dict(orient='records')})  # 將處理過的資料框架轉換為字典並加入到all_tables列表中
    return {"PDF Tables": all_tables}

# 使用範例
pdf_path = '11_R-30iA_HandlingTool基本操作教育訓練教材(NEW).pdf'
result = extract_tables_from_pdf(pdf_path)

print(result)
```

# 合併程式 文本抓取及表格轉字典

```python
import fitz  # PyMuPDF
import pdfplumber
import pandas as pd

def extract_text_from_pdf(pdf_path):
    """
    從PDF檔案中提取文本的函數。
    參數:
        pdf_path (str): PDF檔案的路徑
    返回:
        str: 提取的文本內容
    """
    pdf_document = fitz.open(pdf_path)  # 開啟指定路徑的PDF檔案

    text = ""  # 初始化一個空字串來儲存提取的文本內容

    # 迭代PDF檔案中的每一頁
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)  # 加載當前頁面
        text += page.get_text()  # 提取當前頁面的文本並追加到text變數中

    pdf_document.close()

    return text

def extract_tables_from_pdf(pdf_path):
    """
    從PDF檔案中提取表格資料的函數。
    參數:
        pdf_path (str): PDF檔案的路徑
    返回:
        dict: 包含所有表格的字典
    """
    all_tables = []  # 初始化一個空的列表以儲存所有頁面的表格資料
    with pdfplumber.open(pdf_path) as pdf:
        # 迭代PDF檔案中的每一頁
        for page_num, page in enumerate(pdf.pages):
            table = page.extract_table()  # 從當前頁面提取表格資料
            if table:  # 如果當前頁面有表格資料
                df = pd.DataFrame(table[1:], columns=table[0])  # 創建資料框架，表頭為表格的第一行
                df = df.replace(r'\n', ' ', regex=True)  # 替換資料框架中的換行符為空格
                all_tables.append({f"Page_{page_num+1}": df.to_dict(orient='records')})  # 將處理過的資料框架轉換為字典並加入到all_tables列表中
    return {"PDF Tables": all_tables}

def extract_text_and_tables_from_pdf(pdf_path):
    """
    從PDF檔案中提取文本和表格資料的函數。
    參數:
        pdf_path (str): PDF檔案的路徑
    返回:
        dict: 包含文本和表格資料的字典
    """
    text = extract_text_from_pdf(pdf_path)
    tables = extract_tables_from_pdf(pdf_path)

    result = {
        "Text": text,
        "Tables": tables
    }
    return result

pdf_path = '11_R-30iA_HandlingTool基本操作教育訓練教材(NEW).pdf'
result = extract_text_and_tables_from_pdf(pdf_path)

print(result["Text"], result["Tables"])
```
