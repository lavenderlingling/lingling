# (一) PDF文本處理

```python
import fitz

def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)

    text = ""

    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text += page.get_text()

    pdf_document.close()

    return text

# 示例：提取 PDF 文件中的文本
pdf_path = "機器人類型.pdf"
text = extract_text_from_pdf(pdf_path)
print(text)
```

# (二) PDF表格處理

```python
import pdfplumber
import pandas as pd

with pdfplumber.open("產品使用說明書.pdf") as pdf:
    all_tables = []
    for page in pdf.pages:
        table = page.extract_table()
        if table:
            df = pd.DataFrame(table[1:], columns=table[0])
            df = df.replace(r'\n', ' ', regex=True)
            all_tables.append(df)
```

```python
# 印出讀取結果
for i in range(len(all_tables)):
    df = pd.DataFrame(all_tables[i])
    print(f"Table {i}:")
    print(df)
    print("\n")
```

# PDF表格轉Markdown

```python
# 使用pdfplumber
import pdfplumber

def extract_tables(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        pages = pdf.pages
        for page in pages:
            # 提取頁面中的所有表格
            page_tables = page.extract_tables()
            for table in page_tables:
                yield table  # 每個表格都是一個列表的列表

def tables_to_markdown(tables):
    markdown = ""
    for table in tables:
        # 添加Markdown表格的表頭
        markdown += "| " + " | ".join(table[0]) + " |\n"
        markdown += "| " + " | ".join(['---']*len(table[0])) + " |\n"
        # 添加表格內容
        for row in table[1:]:
            markdown += "| " + " | ".join(row) + " |\n"
        markdown += "\n"  # 添加一個空行在每個表格後
    return markdown

# 你的PDF文件路徑
pdf_path = '機器人類型.pdf'

# 提取PDF中的所有表格
tables = list(extract_tables(pdf_path))

# 將提取的表格轉換為Markdown格式
markdown_tables = tables_to_markdown(tables)
print(markdown_tables)
```
